```python
for index, row in df.iterrows():
    company_name = row['entity name']
    receiver_email = row['email']

    # Change Sender Name！！！！！！！！
    body = generate_email_body_itsuki(company_name)

    # 构建邮件内容
    msg = MIMEText(body, "plain", "utf-8")
    msg['Subject'] = Header(f"TikTok Shopのご案内 – 詳細説明の機会を頂けますでしょうか", "utf-8")
    msg['From'] = sender_email
    msg['To'] = receiver_email
    msg['Bcc'] = ", ".join(bcc_emails)

    to_addrs = [receiver_email] + bcc_emails

    try:
        server.sendmail(sender_email, to_addrs, msg.as_string())
        print(f"✅ 已发送至 {company_name} ({receiver_email})")
        time.sleep(2)  # 控制频率，防止过快
    except Exception as e:
        print(f"❌ 发送失败：{company_name} ({receiver_email})，错误：{e}")

server.quit()
```
